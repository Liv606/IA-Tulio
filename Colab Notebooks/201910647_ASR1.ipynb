```python
from random import choice
import string
import random

string.ascii_letters
string.digits

tam = 0
letras = 0
num = 0

while(tam < 8):
  tam = int(input("Digite o tamanho da senha(mais que 8 caracteres): "))
  if(tam < 8):
    print("A senha deve ter mais que 8 caracteres!")

while(letras + num != tam):
  letras = int(input("Digite a quantidade de letras: "))
  num = int(input("Digite a quantidade de números: "))

  if ((letras + num) != tam):
    print("A soma de letras e números tem que ser igual ao tamanho da senha!")

  else:

    let = string.ascii_letters
    nume = string.digits
    senha = ''
    senha2 = ''
    senhaf = ''
    for i in range(letras):
      senha += choice(let)

    for i in range(num):
      senha2 += choice(nume)

    senhaf = senha + senha2

senhafinal = "".join(random.sample(senhaf, len(senhaf)))

print(senhafinal)
```
